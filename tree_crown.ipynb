```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import scipy.ndimage as ndimage
import scipy.ndimage.filters as filters
from skimage.feature import peak_local_max
```

```python
img = cv2.imread('middle_layer/585vm.tif')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

```python
morph_open_image = ndimage.grey_opening(img, size=(1,2),)
plt.imshow(morph_open_image)
```

```python
morph_dilation_image = ndimage.grey_dilation(morph_open_image,(3,3))
plt.imshow(morph_dilation_image)
```

```python
gass_image = filters.gaussian_filter(morph_dilation_image,sigma=2)
plt.imshow(gass_image)
```

```python
coordinates = peak_local_max(gass_image, min_distance=11)
plt.imshow(gass_image)
plt.plot(coordinates[:, 1], coordinates[:, 0], 'r.')
plt.show()
print(coordinates.shape)
```
